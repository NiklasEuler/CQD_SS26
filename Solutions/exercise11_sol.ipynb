# The problem

Light propagating through an atomic medium is usually scattered and thus absorbed by near-resonant coupling to a transition from the ground ($|g_1\rangle$) to an excited ($|e\rangle$) electronic state. If a second light field ("coupling laser") couples the excited state to a third, long-lived state ($|g_2\rangle$), the medium can again become transparent for the first light field ("probe laser"). This effect, called electromagnetically induced transparency (EIT), has led to a plethora of applications, see e.g. Rev. Mod. Phys. 77, 633 (2005) [1]. The dynamics of the atomic populations can be simulated using the Lindblad master equation, accounting for the coherent driving laser fields and incoherent processes such as spontaneous emission and dephasing, e.g. due to finite laser linewidth. The complex refractive index of the atomic medium is given by $n=\sqrt{1+\chi}\approx 1 + (\chi_R + \chi_I)/2$, where the real and imaginary part of the complex susceptibility $\chi$ account for diffraction and absorption of the probe light respectively. $\chi$ depends on the polarizability of the atomic dipoles in the medium and is thus proportional to expectation value of the operator $|g_1\rangle \langle e|$, i.e. $\rm{tr}[|g_1\rangle \langle e| \rho] = \langle e| \rho |g_1\rangle = \rho_{eg_1}$. This coherence will thus be the main object of interest.

The interaction with the laser fields is described by the Hamiltonian (in rotating wave approximation, in the basis $\{|g_1\rangle,|g_2\rangle,|e\rangle\}$, following the sign convention used in [1])
$$
H=\left[
\begin{array}{ccc}
0 & 0 & -\Omega_p/2 \\
0 & \Delta_p-\Delta_c & -\Omega_c/2 \\
-\Omega_p/2 & -\Omega_c/2 & \Delta_p
\end{array}
\right]
$$

where $\Omega$'s are Rabi-frequencies and $\Delta$'s are the detunings of the lasers from the respective atomic transitions. $\Delta_p-\Delta_c$ is the "two-photon-detuning" sometimes abbreviated as $\delta_2$.

The dissipative processes we want to consider are the spontaneous emission from $e$ to both stable states $g_1$ and $g_2$ and later also the decay from $g_2$ to $g_1$, which destroys the EIT effect. These are accounted for by the jump operators
$$
\Gamma_{eg_1} = \sqrt{\gamma_p} |g_1\rangle \langle e|, \quad \Gamma_{eg_2} = \sqrt{\gamma_c} |g_2\rangle \langle e|, \quad\Gamma_{g_2g_1} = \sqrt{\gamma_g} |g_1\rangle \langle g_2| .
$$
The resulting master equation reads
$$
\dot\rho = -i[H,\rho] + \mathcal{L}_{\Gamma_{eg_1}} [\rho]+ \mathcal{L}_{\Gamma_{eg_2}} [\rho]+ \mathcal{L}_{\Gamma_{g_2g_1}} [\rho]
$$
with the Lindblad superoperator
$$
\mathcal{L}_{\Gamma_x} [\rho] = \Gamma_x \rho \Gamma_x^\dagger - \frac{1}{2}(\Gamma_x^\dagger\Gamma_x \rho + \rho \Gamma_x^\dagger\Gamma_x)
$$

### Exercise 1: System setup and steady-state calculation

Construct the Liouvillian superoperator in matrix form and test your implementation by calculating the steady state for the case of zero two-photon detuning ($\delta_2=0$) and $\gamma_g=0$, which is the pure state ("dark state")
$$
|d\rangle = \frac{\Omega_c |g_1\rangle - \Omega_p |g_2\rangle}{\sqrt{\Omega_p^2 + \Omega_c^2}}.
$$

Detailed instructions (you can of course proceed differently if you want. I am pretty sure the method described below is not the most elegant one...):
- Building the Liouvillian in matrix form: We want to formulate the master equation in the standard form of an ordinary differential equation $\dot\rho = \mathcal{L} \rho$ with $\rho$ the density matrix in vector form and $\mathcal{L}$ a matrix summarizing the right-hand side of the master equation. However, the master equation is more conveniently formulated using dot-products between matrices representing operators in Hilbert space. Therefore, it makes sense to write a function that takes the vectorized density matrix as an input (along with the Hamiltonian matrix and the jump operators used in the Lindblad terms), converts it into a matrix using reshape(), calculates the right-hand side of the master equation, and returns the result in vector form by using reshape again. To build the Liouvillian matrix you can loop over all the projectors $|i\rangle\langle j|$ and pass them to this function. Those projectors are the basis vectors spanning the space of vectorized density matrices. The outputs will be the columns of the Liouvillian matrix.
- Calculating steady states: The steady state is the unique state for which the right-hand side of the master equation gives zero. Calculating this state numerically is less trivial than it seems. If we just try solving the linear system of equations $\mathcal{L}\rho = 0$ it won't work since $\rho=0$ is always a trivial solution. We need to ensure that the trace of the density matrix is 1. This can be achieved by dropping the first line of the master equation ($\dot\rho_{g_1g_1}=...$) and replacing $\rho_{g_1g_1} = 1-\rho_{g_2g_2}-\rho_{ee}$ everywhere. The resulting inhomogeneous system of 8 equations can be solved using np.linalg.solve(). $\rho_{g_1g_1}$ can be prepended to the solution by using the trace condition again. There are many other ways to calculate steady states; see for example the QuTiP documentation (http://qutip.org/docs/4.0.2/guide/guide-steady.html). You can of course also use QuTiP routines to check your results.

```python
# parameters
params_EIT_test = {
    "omegaP": 1.0,
    "omegaC": 1.0,
    "gammaP": 1.0,
    "gammaC": 0.0,
    "Gamma": 0.0,
    "DeltaP": 0.0,
    "DeltaC": 0.0
}

# calculate steady state
Lmat = cqd.open_systems.build_L_mat_EIT(params_EIT_test)
ss_mat = cqd.open_systems.rho_ss(Lmat)
# print(Lmat.real)
# print(Lmat.imag)
print("Steady state matrix:")
print(ss_mat)

### testing: ###

# check that RHS of ME is zero for ss
ssVec = ss_mat.reshape((9,))
print("Steady state vector:")
print(ssVec)
print("RHS of Master Equation (should be zero):")
print(Lmat @ ssVec)

# Manually build the Liouvillian matrix and check that it is the same as the one built by the function
# build the Hamiltonian
H = np.array([[0, 0, -params_EIT_test["omegaP"] / 2], [0, params_EIT_test["DeltaP"] - params_EIT_test["DeltaC"], -params_EIT_test["omegaC"] / 2], [-params_EIT_test["omegaP"] / 2, -params_EIT_test["omegaC"] / 2, params_EIT_test["DeltaP"]]])
dimH = len(H)
# build jump operators
Lp = np.sqrt(params_EIT_test["gammaP"]) * cqd.open_systems.pij(dimH, 0, 2)
Lc = np.sqrt(params_EIT_test["gammaC"]) * cqd.open_systems.pij(dimH, 1, 2)
Lg = np.sqrt(params_EIT_test["Gamma"]) * cqd.open_systems.pij(dimH, 0, 1)
Llist = np.array([Lp, Lc, Lg])

print("Manually calculated RHS of Master Equation (should be zero):")
print(cqd.open_systems.ME_RHS(ssVec, H, Llist))

normFact = params_EIT_test["omegaP"]**2 + params_EIT_test["omegaC"]**2
print("Analytical values for comparison (rhog1g1, rhog1g2, rhog2g2):")
print(params_EIT_test["omegaC"] **2 / normFact, params_EIT_test["omegaP"] **2 / normFact, -params_EIT_test["omegaC"] * params_EIT_test["omegaP"] / normFact)
```

### Exercise 3: Dissipative dynamics

Solve the time-dependent master equation with initial state $\rho_0 = |g_1\rangle \langle g_1|$. You can use a scipy integrator, for example the one used in exercise sheet 5. Plot the populations of the levels and the real and imaginary parts of $\rho_{eg_1}$ as a function of time. A good parameter choice is $\Omega_p = 1$, $\Omega_c = \gamma_p = 2$, with all other parameters set to zero, integrating up to $t=20$. Check qualitatively how the relaxation time depends on the parameters, especially on $\Omega_c$ and $\gamma_p$.

```python
# parameters
params_EIT_dynamics = {
    "omegaP": 1.0,
    "omegaC": 2.0,
    "gammaP": 2.0,
    "gammaC": 0.0,
    "Gamma": 0.0,
    "DeltaP": 0.0,
    "DeltaC": 0.0
}

# initial condition
ini = cqd.utility.pij(3, 0, 0).reshape(9,) # all in g1

# time grid
tend = 20
dt = .1
tsteps = int(tend / dt)
tvec = cqd.utility.create_tvecs(tsteps, dt)

# container for results
rho_t = np.zeros((len(tvec), 3, 3),dtype='complex')

# build liouvillian
L_mat = cqd.open_systems.build_L_mat_EIT(params_EIT_dynamics)
# print(L_mat.real)
# print(L_mat.imag)

# initialize the integrator
y0, t0 = ini, 0
# print(y0)

r = ode(cqd.integrators.linblad_master_eq).set_integrator('zvode', method='adams', with_jacobian=False)
r.set_initial_value(y0, t0).set_f_params(L_mat)

# store initial density matrix
rho_t[0] = ini.reshape((3, 3))

# integration
i = 1
while r.successful() and r.t < tend and i < len(tvec):
    r.integrate(r.t + dt)
    rho = r.y
    rho_t[i] = np.reshape(rho, (3, 3))
    i += 1
    
# steady state
ss = cqd.open_systems.rho_ss(L_mat)
```

```python
# plotting
tab10 = plt.get_cmap('tab10')  # get the tab10 colormap for consistent colors
# populations
plt.plot(tvec, np.real(rho_t[:, 0, 0]), label='g_1')
plt.plot(tvec, np.real(rho_t[:, 1, 1]), label='g_2')
plt.plot(tvec, np.real(rho_t[:, 2, 2]), label='e')
plt.plot([0, tend], [np.real(ss[0, 0]), np.real(ss[0, 0])], ls='--', color=tab10(0), label='ss g_1')
plt.plot([0, tend], [np.real(ss[1, 1]), np.real(ss[1, 1])], ls='--', color=tab10(1), label='ss g_2')
plt.plot([0, tend], [np.real(ss[2, 2]), np.real(ss[2, 2])], ls='--', color=tab10(2), label='ss e')
plt.xlabel('$t$')
plt.ylabel('populations')
plt.legend(ncol=2)
plt.show()

# absorption
plt.plot(tvec, np.imag(rho_t[:, 2, 0]), color=tab10(3), label='absorption')
plt.plot([0, tend], [np.imag(ss[2, 0]), np.imag(ss[2, 0])], ls='--', color=tab10(3), label='ss absorption')
plt.xlabel('$t$')
plt.ylabel('$\\propto $Im$[\\chi]$')
plt.legend()
plt.show()

# diffraction
plt.plot(tvec, np.real(rho_t[:, 2, 0]), color = tab10(3), label='diffraction')
plt.plot([0, tend], [np.real(ss[2, 0]), np.real(ss[2, 0])], ls='--', color=tab10(3), label='ss diffraction')
plt.xlabel('$t$')
plt.ylabel('$\\propto $Re$[\\chi]$')
plt.legend()
plt.show()
```

Observations:
- The system relaxes into the dark state.
- The relaxation timescale depends on $\Omega_c$.
- Coherent oscillations appear when $\Omega_c/\gamma_p$ is not too small.

```python

H, L_list = cqd.open_systems.build_EIT_operators(params_MCWF_standard_EIT)
H_non_herm = cqd.open_systems.build_H_non_herm(H, L_list)

ini = np.array([1, 0, 0], dtype='complex')
dt = 0.05
tend = 10
tsteps = int(tend / dt)
tvec = cqd.utility.create_tvecs(tsteps, dt)

n_traj = 300

all_trajectories = np.zeros((n_traj, len(tvec), len(H_non_herm)), dtype='complex')

for traj_idx in range(n_traj):
    all_trajectories[traj_idx] = cqd.open_systems.get_trajectory(H_non_herm, L_list, ini, tvec)
    if traj_idx % 50 == 0:
        print(traj_idx, end=' ')
        
for i in range(20):
    plt.plot(tvec, np.abs(all_trajectories[i, :, 0])**2)
plt.xlabel('$t$')
plt.ylabel('population')
# plt.legend(['g_1','g_2','e'])
plt.title('single trajectories')
plt.show()

plt.plot(tvec, np.mean(np.abs(all_trajectories[:, :, 0]) ** 2, axis=0))
plt.plot(tvec, np.mean(np.abs(all_trajectories[:, :, 1]) ** 2, axis=0))
plt.plot(tvec, np.mean(np.abs( all_trajectories[:, :, 2]) ** 2, axis=0))
plt.xlabel('$t$')
plt.ylabel('populations')
plt.legend(['g_1', 'g_2', 'e'])
plt.title('averages over trajectories')
plt.show()
# Integrate master equation for comparison

# initial condition
ini = cqd.utility.pij(3, 0, 0).reshape(9,) # all in g1

# time grid
#tend = 10 # same as above
dt = .1
tsteps = int(tend/dt)
tvec_ME = cqd.utility.create_tvecs(tsteps, dt)

# container for results
rho_t = np.zeros((len(tvec_ME), 3, 3), dtype='complex')

# build Lindbladian
L_mat = cqd.open_systems.build_L_mat_EIT(params_MCWF_standard_EIT)

# initialize the integrator
y0, t0 = ini, 0
r = ode(cqd.integrators.linblad_master_eq).set_integrator('zvode', method='adams', with_jacobian=False)
r.set_initial_value(y0, t0).set_f_params(L_mat)

# store initial density matrix
rho_t[0] = ini.reshape((3, 3))

# integration
i=1
while r.successful() and r.t < tend and i < len(tvec_ME):
    r.integrate(r.t + dt)
    rho = r.y
    rho_t[i] = np.reshape(rho, (3, 3))
    i += 1
    
# steady state
ss = cqd.open_systems.rho_ss(L_mat)

# plotting
tab10 = plt.get_cmap('tab10')  # get the tab10 colormap for consistent colors
# populations
plt.plot(tvec_ME, np.real(rho_t[:, 0, 0]), color=tab10(0), label='g_1')
plt.plot(tvec_ME, np.real(rho_t[:, 1, 1]), color=tab10(1), label='g_2')
plt.plot(tvec_ME, np.real(rho_t[:, 2, 2]), color=tab10(2), label='e')
plt.plot([0, tend], [np.real(ss[0, 0]), np.real(ss[0, 0])], ls='--', color=tab10(0), label='ss g_1')
plt.plot([0, tend], [np.real(ss[1, 1]), np.real(ss[1, 1])], ls='--', color=tab10(1), label='ss g_2')
plt.plot([0, tend], [np.real(ss[2, 2]), np.real(ss[2, 2])], ls='--', color=tab10(2), label='ss e')
plt.xlabel('$t$')
plt.ylabel('populations')
# mc-wf results
plt.plot(tvec, np.mean(np.abs(all_trajectories[:, : ,0]) ** 2, axis=0), ls=':', color=tab10(0), label='MC-WF g_1')
plt.plot(tvec, np.mean(np.abs(all_trajectories[:, :, 1]) ** 2, axis=0), ls=':', color=tab10(1), label='MC-WF g_2')
plt.plot(tvec, np.mean(np.abs(all_trajectories[:, :, 2]) ** 2, axis=0), ls=':', color=tab10(2), label='MC-WF e')
plt.legend(ncol=3)
plt.show()

# absorption
chi_Im = np.mean(all_trajectories[:, :, 0].conjugate() * all_trajectories[:, :, 2], axis=0).imag
plt.plot(tvec_ME, np.imag(rho_t[:, 2, 0]), color=tab10(3), label='ME absorption')
plt.plot([0, tend], [np.imag(ss[2, 0]), np.imag(ss[2, 0])], ls='--', color=tab10(3), label='ss absorption')
plt.plot(tvec, chi_Im, ls=':', color=tab10(3), label='MC-WF absorption')
plt.xlabel('$t$')
plt.ylabel('$\\propto $Im$[\\chi]$')
plt.legend()
plt.show()

# diffraction
chiRe = np.mean(all_trajectories[:, :, 0].conjugate() * all_trajectories[:, :, 2], axis=0).real
plt.plot(tvec_ME, np.real(rho_t[:, 2, 0]), color=tab10(3), label='ME diffraction')
plt.plot([0, tend], [np.real(ss[2, 0]), np.real(ss[2, 0])], ls='--', color=tab10(3), label='ss diffraction')
plt.plot(tvec, chiRe, ls=':', color=tab10(3), label='MC-WF diffraction')
plt.xlabel('$t$')
plt.ylabel('$\\propto $Re$[\\chi]$')
plt.legend()
plt.show()
```
